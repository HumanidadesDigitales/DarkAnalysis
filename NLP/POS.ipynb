```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import spacy
from collections import Counter
from string import punctuation
from spacy import displacy
```

```python
def reading(read_path,save_path=None):

    df = pd.read_csv(read_path, sep=",")

    return df

df = reading("/Users/roicort/Documents/GitHub/DarkMagic/NLP/corpus.csv")
df.rename(columns={'Dónde se compró': 'Procedencia', 'Datos adicionales': 'Información'}, inplace=True)
df.drop(columns=["Otras versiones"],inplace=True)
df.dropna(subset=['Oración'],inplace=True)
```

```python
df
```

```python
nlp = spacy.load("es_core_news_sm")
```

```python
text = df.iloc[12]["Oración"].replace("\n"," ")
print(text)
doc = nlp(text)
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,token.shape_, token.is_alpha, token.is_stop)
```

```python
sentence_spans = list(doc.sents)
displacy.render(sentence_spans, style="dep")
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```
