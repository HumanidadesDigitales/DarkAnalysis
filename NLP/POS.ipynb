# Construyendo un grafo

```python
spells
```

```python
nlp = spacy.load("es_core_news_md")
import unidecode
```

```python
entitiesdict = {}
for _, row in spells.iterrows():
    text = row["Oración"].replace("\n"," ")
    #print(text)
    doc = nlp(text)
    entities = [unidecode.unidecode(ent.text.lower().replace("_","")) for ent in doc.ents]
    #print(entities)
    ID = row["Id"]
    entitiesdict[ID]=entities
```

```python
admat = pd.DataFrame(columns = entitiesdict.keys(),index=entitiesdict.keys())

import networkx as nx
G = nx.Graph()

for k in entitiesdict.keys():
    for j in entitiesdict.keys():
        if k!=j:
            a = entitiesdict[k]
            b = entitiesdict[j]
            inter = set(a).intersection(set(b))
            admat[k][j] = list(inter)
            if admat[k][j] != []:
                G.add_node(k,ents=a)
                G.add_node(j,ents=b)
                G.add_edge(k,j,weight=len(admat[k][j]))
admat
```

### Dibujando el grafo

```python

pos = nx.kamada_kawai_layout(G)

nx.draw_networkx_nodes(G,pos,node_size=10,node_color="red",alpha=0.5)
nx.draw_networkx_edges(G,pos,edge_color="black",alpha=0.1)
nx.draw_networkx_labels(G,pos,font_size=2)

plt.show()
```

### Buscando comunidades en el grafo

```python
import networkx.algorithms.community as nxcom
from wordcloud import WordCloud

communities = sorted(nxcom.greedy_modularity_communities(G), key=len, reverse=True)
print(len(communities))
```

### Caracterizando cada comunidad encontrada

```python
spacy_stopwords = nlp.Defaults.stop_words

for c in communities:
    print(c)
    biglist = []
    for e in list(c):
        biglist+=G.nodes[e]["ents"]
    unique_string=(" ").join(biglist)
    without_stopwords = (" ").join([word for word in unique_string.split() if word not in spacy_stopwords]).replace("oracion","").replace("y","").replace(".","")
    #print(without_stopwords)
    wordcloud = WordCloud(width = 2560, height = 1600,background_color ='white', min_font_size = 10).generate(without_stopwords) 
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    plt.show() 
```
